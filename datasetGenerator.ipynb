````python
import pandas as pd
# create an empty dataframe with only one column named "text"
df = pd.DataFrame(columns=["text"])
````

````python
# make a function to read a string in a variable and concatenate it to the dataframe
def add_to_df(text):
    global df
    df = pd.concat([df, pd.DataFrame([text], columns=["text"])], ignore_index=True)
````

````python
#get openai api key from .env file
import os
from dotenv import load_dotenv
import openai
load_dotenv()
openai.api_key = os.getenv('OPENAI_API_KEY')
````

````python

def get_completion(prompt, model="gpt-3.5-turbo"):
    messages = [{"role": "user", "content": prompt}]
    response = openai.ChatCompletion.create(
        model=model,
        messages=messages,
        temperature=0, # this is the degree of randomness of the model's output
    )
    return response.choices[0].message["content"]
````

````python
# a function that make the prompt from two lists of first names and last names
def makePrompt(first_names,last_names):
    prompt = f"""
    Your task is to a short text in French where first names and last names are followed by the mention <first name> and <last name> respectively.
    Here is an example of what the output should look like. The topic, tone, and style of the text should be different from the example below:
    ```
    Voici une histoire passionnante sur un jeune garçon nommé Pierre <prénom>. Il vivait dans un petit village appelé Saint-Étienne <nom de famille>. Pierre <prénom> aimait explorer la nature et passer du temps avec ses amis. Un jour, il rencontra une fille charmante nommée Sophie <prénom>. Ils devinrent rapidement les meilleurs amis. Pierre <prénom> et Sophie <prénom> passaient leurs journées à jouer dans les champs et à rire ensemble. Cependant, un jour, Pierre <prénom> tomba malade et Sophie <prénom> fut très inquiète. Heureusement, grâce aux soins attentifs de sa famille et des médecins, Pierre <prénom> se rétablit rapidement. Sophie <prénom> était soulagée de voir son ami en bonne santé. Depuis ce jour, Pierre <prénom> et Sophie <prénom> étaient inséparables et partageaient de nombreux moments de bonheur ensemble.
    ```
    The names should be as varied as possible, including foreign names. The topic could be about anything, but it should be different from the example above.
    the first names should be different from the following list: {first_names}
    the last names should be different from the following list: {last_names}
    Here are a few suggestions of topics: judiciaire, politique, sport, économie, santé, culture, technologie, environnement, éducation, société, international, sciences, médias, religion, faits divers, people, etc.
    Here are few suggestions of styles: informatif, narratif, descriptif, argumentatif, explicatif, injonctif, poétique, humoristique, etc.
    The output should be a JSON object with the following keys: text, first_names, last_names.
    """
    return prompt

````

````python
import json
first_names = ["Pierre"]
last_names = ["Saint-Étienne"]
````

````python

# add texts in a loop to dataframe
for i in range(10):
    prompt = makePrompt(first_names,last_names)
    response = get_completion(prompt)
    print(response)
    response = json.loads(response)
    text = response["text"]
    print(text)
    # get the contents of the "first_names" and "last_names" keys from the JSON object in the response and append them to the lists
    first_names = first_names + response["first_names"]
    last_names = last_names + response["last_names"]
    print(first_names)
    print(last_names)


    add_to_df(text)

df.head()
````
