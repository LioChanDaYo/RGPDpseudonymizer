````python
import pandas as pd
# create an empty dataframe with only one column named "text"
df = pd.DataFrame(columns=["text"])
````

````python
# make a function to read a string in a variable and concatenate it to the dataframe
def add_to_df(text):
    global df
    df = pd.concat([df, pd.DataFrame([text], columns=["text"])], ignore_index=True)
````

````python
#get openai api key from .env file
import os
from dotenv import load_dotenv
import openai
load_dotenv()
openai.api_key = os.getenv('OPENAI_API_KEY')
````

````python

def get_completion(prompt, model="gpt-3.5-turbo"):
    messages = [{"role": "user", "content": prompt}]
    response = openai.ChatCompletion.create(
        model=model,
        messages=messages,
        temperature=0.8, # this is the degree of randomness of the model's output
    )
    return response.choices[0].message["content"]
````

````python
def makePrompt(first_names,last_names,topic,beginning):
    prompt = f"""
    Your task is to write a short text in French where first names and last names are followed by the mention <first name> and <last name> respectively.
    the first names should NOT be any element from the following list: {first_names}
    the last names should NOT be any element from the following list: {last_names}
    the topic should NOT be any element from the following list: {topic}
    the beginning should NOT be any element from the following list: {beginning}
    The names should be as varied as possible, including foreign names. The topic could be about anything, but it should be different from the example below.
    The output should ONLY be a JSON object with the following keys: text, first_names, last_names, topics, beginnings. Each key except text should be a list of strings from the text. The text should be a string.
    Here is an example of what the text in the output should look like. The topic, tone, and literary genre of the text should be different from the example below:
    ```
    Voici une histoire passionnante sur un jeune garçon nommé Pierre <first name>. Il vivait dans un petit village appelé Saint-Étienne. Pierre <first name> Leblanc <last name> aimait explorer la nature et passer du temps avec ses amis. Un jour, il rencontra une fille charmante nommée Sophie <first name>. Ils devinrent rapidement les meilleurs amis. Pierre <first name> et Sophie <first name> passaient leurs journées à jouer dans les champs et à rire ensemble. Cependant, un jour, Pierre <first name> tomba malade et Sophie <first name> fut très inquiète. Heureusement, grâce aux soins attentifs de sa famille et des médecins, Pierre <first name> se rétablit rapidement. Sophie <first name> était soulagée de voir son ami en bonne santé. Depuis ce jour, Pierre <first name> et Sophie <first name> étaient inséparables et partageaient de nombreux moments de bonheur ensemble.
    ```
    """
    return prompt
````

````python
import json
def is_json(myjson):
  try:
    json.loads(myjson)
  except ValueError as e:
    return False
  return True

first_names = ["Pierre","Sophie"]
last_names = ["Leblanc"]
topic = []
beginning = ["Voici une histoire"]
````

````python

# add texts in a loop to dataframe
for i in range(10):
    prompt = makePrompt(first_names,last_names,topic,beginning)
    response = get_completion(prompt)
    print("réponse complète : " + response)
    if is_json(response):
        response = json.loads(response)
        text = response["text"]
        print("le texte est : " + text)
        # get the contents of the "first_names" and "last_names" keys from the JSON object in the response and append them to the lists
        first_names = first_names + response["first_names"]
        # if first_names is longer than 15 elements, keep only the last 15 elements
        if len(first_names) > 15:
            first_names = first_names[-15:]
        last_names = last_names + response["last_names"]
        if len(last_names) > 15:
            last_names = last_names[-15:]
        topic = topic + response["topics"]
        if len(topic) > 15:
            topic = topic[-15:]
        beginning = beginning + response["beginnings"]
        if len(beginning) > 15:
            beginning = beginning[-15:]
        print("les prénoms sont : ")
        print(first_names)
        print("les noms de famille sont : ")
        print(last_names)
        print("les sujets sont : ")
        print(topic)
        print("les débuts sont : ")
        print(beginning)
    else:
        text = response
        print(text)
    add_to_df(text)

````
